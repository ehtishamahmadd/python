# Dealing with missing values
- In a dataset missing values are either ? or NaN, or 0 or a blank cell.
- Jab kabhi data na ho kisi aik row main kisi bhi aik parameter ka.(missing row value in a parameter)
> Steps:
1. Try to collect again data and see if there is any mistake.
2. Remove the column or row cell if possible(doesnt effects the data)
3. Replace the missing values
   1. How?
      1. Average value of entire variable(put mean of a column in a specific row cell) or similar data point
      2. frequency or MODE replacement
      3. Replace based on other functions(Data Sampler knows that) (same value like other values)
      4. Machine Learning algorithm can also be used to find missing values(through linear regression)
      5. Leave it like as it is
   2. Why?
      1. Its better because no data is lost
      2. Less accurate

# Data Normalization
- Uniform the data
- Making sure they have same impact
- Aik machli samundar main aur aik jar main(Example: not good comparison)
- Also for computational reasons

```python
kashti.head()
```

```python
ks4=ks1[['age in days','fare']]
ks4.head()
```

- The above data is really wide range and we need to normalize and hard to compare
- Normalization change the values to the range 0-1 (now both variable has similar influence on our models

## Method of Normalization
1. Simple feature scaling
   1. x(new)=x(old)/x(max)
2. Min-Max method
3. Z-score (standard score) -3 to +3
4. Log transformation

```python
# simple feature scaling
ks4['fare']=ks4['fare']/ks4['fare'].max()
ks4.head()
```

```python
# simple feature scaling
ks4['age in days']=ks4['age in days']/ks4['age in days'].max()
ks4.head()
```

```python
# Min - Max method
ks4['fare']=(ks4['fare']-ks4['fare'].min())/(ks4['fare'].max()-ks4['fare'].min())
ks4.head()
```

```python
# Z-score method(standard score)
ks4['fare'] = (ks4['fare']-ks4['fare'].mean())/ks4['fare'].std()
ks4.head()
```

```python
# log transformation(range of value is same but it jsut concises the values)
ks['fare'] = np.log(ks['fare'])
ks.head()
```
