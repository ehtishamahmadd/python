# Logistic Classification and Image Analysis

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# import online
from sklearn.datasets import load_digits
digits= load_digits()
```

```python
# input variables ya features or X
digits.data.shape
X=digits.data

#1797 total rows(pictures) and 64 is size(8*8)
```

```python
# output labels or y
digits.target.shape
y= digits.target
# only labels no images here ( because no size is here)
```

```python
import seaborn as sns
sns.set_style('ticks')
plt.figure(figsize=(20,4))
for index,(image, label) in enumerate(zip(digits.data[0:10], digits.target[0:10])):
    plt.subplot(1,10, index+1)
    plt.imshow(np.reshape(image,(8,8)), cmap=plt.cm.gray)
    plt.title('Training: %i\n' % label, fontsize= 20)
```

```python
plt.figure(figsize=(20,4))
for index,(image, label) in enumerate(zip(digits.data[0:20], digits.target[0:20])):
    plt.subplot(1,20, index+1)
    plt.imshow(np.reshape(image,(8,8)), cmap=plt.cm.gray)
    plt.title(label, fontsize= 20)
```

```python
#help(plt)

#pyplot manual
```

```python
# split the data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.25, random_state=0)
```

```python
print('train input data: ', X_train.shape)
print('test input data: ', X_test.shape)
print('train input data: ', y_train.shape)
print('test input data: ', y_test.shape)
```

```python
# training model

from sklearn.linear_model import LogisticRegression
model = LogisticRegression().fit(X_train, y_train)
model
```

```python
# prediction
model.predict(X_test[0:10])
prediction= model.predict(X_test)
prediction
# extracting 0 to 10 rows
```

```python
# accuracy test

score = model.score(X_test,y_test) #x_test is the predicted value and compared with original y values
print("accuracy score = ", score)
```

```python
# confusion matrix
from sklearn import metrics
cm= metrics.confusion_matrix(y_test, prediction)
cm
# the other values except the diagonal show that machine is confused in values.
 
```

```python
plt.figure(figsize=(9,9))
sns.heatmap(cm, annot= True, fmt='.3f', linewidth=.5, square=True,
        cmap='Spectral')
plt.ylabel('Actual Output')
plt.xlabel('Predicted Output')
all_sample_title='Acuracy Score: {0}'.format(score)
plt.title(all_sample_title, size= 15)
```
