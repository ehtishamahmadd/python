```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

sns.set_theme(style="ticks")
palette = sns.color_palette("rocket_r")
fao = pd.read_csv("fertilizer_consumption.csv")

sns.relplot(
    data=fao,
    x="Year", y="Value",
    hue="Item",
    kind="line",
    height=5, aspect=1.5, facet_kws=dict(sharex=False),

)
plt.title("Fertilizers consumption in nutrients in PAKISTAN",size=15)
plt.ylabel("Value in Tonnes")
```

```python
import seaborn as sns
sns.set_theme(style="whitegrid")

fao = pd.read_csv("fertilizer_consumption.csv")

# Draw a nested violinplot and split the violins for easier comparison
sns.violinplot(data=fao, x="Item Code", y="Value", hue="Item",
              linewidth=1)
sns.despine(left=True)
```

```python
import seaborn as sns
import pandas as pd
sns.set_theme(style="whitegrid")

fao = pd.read_csv("fertilizer_consumption.csv")

cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
g = sns.relplot(
    data=fao,
    y="Year", x="Value",
    hue="Item Code", size="Item",
    palette=cmap, sizes=(10, 200),
)
g.set(xscale="log", yscale="log")
g.ax.xaxis.grid(True, "minor", linewidth=.25)
g.ax.yaxis.grid(True, "minor", linewidth=.25)
g.despine(left=True, bottom=True)
```
